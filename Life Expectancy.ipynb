```python
# 1.3 Missing values (counts + percent)
missing_count = df.isnull().sum().sort_values(ascending=False)
missing_pct   = (df.isnull().mean()*100).sort_values(ascending=False)

missing_summary = pd.concat([missing_count, missing_pct], axis=1, keys=['missing_count','missing_pct'])
display(missing_summary.head(30))   # top 30 with most missing
```

```python
# 1.4 Check & remove duplicate rows

print("Duplicate rows before:", df.duplicated().sum())

# If you are OK to drop exact duplicate rows, run:
df = df.drop_duplicates().reset_index(drop=True)

print("Duplicate rows after:", df.duplicated().sum())
```

```python
# 1.5 Convert object columns that are mostly numeric to numeric type

obj_cols = df.select_dtypes(include=['object']).columns.tolist()
auto_converted = []

for c in obj_cols:
    # remove commas and percent signs then try numeric conversion
    cleaned = df[c].astype(str).str.replace(',', '').str.replace('%','')
    coerced = pd.to_numeric(cleaned, errors='coerce')
    frac_numeric = coerced.notnull().mean()  # fraction parseable as number

    if frac_numeric > 0.8:   # if >80% parseable as numeric, convert the column
        df.loc[:, c] = coerced
        auto_converted.append(c)

if auto_converted:
    print("Auto-converted these object cols to numeric:", auto_converted)
else:
    print("No object columns looked numeric enough to auto-convert.")
```

```python
# 1.6 Impute missing values (numeric -> median, categorical -> most frequent)

from sklearn.impute import SimpleImputer
import numpy as np

# Identify numeric and categorical columns robustly

numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
cat_cols     = df.select_dtypes(include=['object','category']).columns.tolist()

# Numeric imputer

if numeric_cols:
    num_imputer = SimpleImputer(strategy='median')
    df.loc[:, numeric_cols] = pd.DataFrame(
        num_imputer.fit_transform(df[numeric_cols]),
        columns=numeric_cols,
        index=df.index
    )

# Categorical imputer (most frequent)

if cat_cols:
    cat_imputer = SimpleImputer(strategy='most_frequent')
    df.loc[:, cat_cols] = pd.DataFrame(
        cat_imputer.fit_transform(df[cat_cols]),
        columns=cat_cols,
        index=df.index
    )

print("Missing values after imputation (total):", df.isnull().sum().sum())
```

```python
# 1.8 Convert categorical columns to numeric
# 1) Binary categorical -> 0/1
binary_cols = [c for c in cat_cols if df[c].nunique()==2]
for c in binary_cols:
    df.loc[:, c] = df[c].astype('category').cat.codes

# 2) Multi-category -> one-hot encoding (drop_first=True avoids perfect multicollinearity)
multi_cat_cols = [c for c in cat_cols if df[c].nunique() > 2]
if multi_cat_cols:
    df = pd.get_dummies(df, columns=multi_cat_cols, drop_first=True)

print("After encoding, shape:", df.shape)

```

# Step 3: Feature Engineering

--- Split dataset into features (X) and target (y).

--- Normalize/scale data if necessary.

--- Perform train-test split.

```python
# 3.1 Split dataset into Features (X) and Target (y). This separates what we use for prediction (X) and what we want to predict (y).

target_col = "Life expectancy "   # target column (we have to notice the space at end!)
X = df.drop(columns=[target_col]) # all features except target
y = df[target_col]                # target

print("Features shape (X):", X.shape)
print("Target shape (y):", y.shape)
```

```python
# 3.2 Normalize / Scale Data
#Tree-based models (Decision Tree, Random Forest) donâ€™t need scaling, but KNN does. Weâ€™ll scale everything for consistency.

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # result is a NumPy array

print("Scaled feature matrix shape:", X_scaled.shape)
```

```python
# 3.3 â€” Train-Test Split
# We will keep 20% of the data for testing. This ensures we train on one part of the data and test on unseen data.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print("Training set:", X_train.shape, y_train.shape)
print("Test set:", X_test.shape, y_test.shape)


```

# Step 4: Model Training :

--- KNN Classifier  

--- Decision Tree Classifier

--- Random Forest Classifier

--- Compare baseline results.

```python
#4.0 â€” Convert Life Expectancy into Categories
#Weâ€™ll split life expectancy into 3 classes: Low (bottom 33%) , Medium (middle 33%), High (top 33%)
# Instead of predicting exact years, the models will predict a class.

import pandas as pd

# Convert target into categories
y_class = pd.qcut(y, q=3, labels=["Low", "Medium", "High"])

print("Class distribution:\n", y_class.value_counts())
```

```python
from sklearn.model_selection import train_test_split

# Train-test split (use y_class, not y)
X_train, X_test, y_train, y_test = train_test_split(
    X, y_class, test_size=0.2, random_state=42, stratify=y_class
)

print("Train shape:", X_train.shape, "| Test shape:", X_test.shape)
print("Classes in y_train:", y_train.unique())
```

```python
#4.1 â€” KNN Classifier

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier()   # default K=5
knn.fit(X_train, y_class.loc[y_train.index])

y_pred_knn = knn.predict(X_test)
print("KNN Accuracy:", accuracy_score(y_class.loc[y_test.index], y_pred_knn))
```

```python
# 4.2 â€” Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_class.loc[y_train.index])

y_pred_dt = dt.predict(X_test)
print("Decision Tree Accuracy:", accuracy_score(y_class.loc[y_test.index], y_pred_dt))
```

```python
# 4.3 â€” Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_class.loc[y_train.index])

y_pred_rf = rf.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_class.loc[y_test.index], y_pred_rf))
```

```python
# 4.4 â€” Compare Results.
# This shows which model performs best at baseline (before tuning).

print("Model Comparison (Baseline):")
print("KNN Accuracy          :", accuracy_score(y_class.loc[y_test.index], y_pred_knn))
print("Decision Tree Accuracy:", accuracy_score(y_class.loc[y_test.index], y_pred_dt))
print("Random Forest Accuracy:", accuracy_score(y_class.loc[y_test.index], y_pred_rf))
```

# 6. Hyperparameter Tuning

Use RandomizedSearchCV to optimize hyperparameters:

--- KNN â†’ n_neighbors, weights, metric.

--- Decision Tree â†’ max_depth, min_samples_split.

--- Random Forest â†’ n_estimators, max_depth, min_samples_split.

--- Compare default vs tuned models.

```python
# Step 6. Hyperparameter Tuning (RandomizedSearchCV)
# Weâ€™ll tune KNN, Decision Tree, and Random Forest using RandomizedSearchCV to improve performance beyond the baseline models.

from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from scipy.stats import randint as sp_randint

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# KNN
knn_params = {
    "n_neighbors": sp_randint(1, 30),
    "weights": ["uniform", "distance"],
    "metric": ["euclidean", "manhattan", "minkowski"]
}
knn_rs = RandomizedSearchCV(
    KNeighborsClassifier(), knn_params, n_iter=20, scoring="f1_macro",
    cv=cv, random_state=42, n_jobs=-1
)
knn_rs.fit(X_train, y_train)
print("Best KNN Params:", knn_rs.best_params_)
print("Best KNN Score (CV F1):", knn_rs.best_score_)

# Decision Tree
dt_params = {
    "max_depth": sp_randint(2, 50),
    "min_samples_split": sp_randint(2, 20),
    "min_samples_leaf": sp_randint(1, 20),
    "criterion": ["gini", "entropy", "log_loss"]
}
dt_rs = RandomizedSearchCV(
    DecisionTreeClassifier(random_state=42), dt_params, n_iter=20,
    scoring="f1_macro", cv=cv, random_state=42, n_jobs=-1
)
dt_rs.fit(X_train, y_train)
print("Best Decision Tree Params:", dt_rs.best_params_)
print("Best Decision Tree Score (CV F1):", dt_rs.best_score_)

# Random Forest
rf_params = {
    "n_estimators": sp_randint(50, 300),
    "max_depth": sp_randint(2, 50),
    "min_samples_split": sp_randint(2, 20),
    "min_samples_leaf": sp_randint(1, 20),
    "max_features": ["sqrt", "log2", None]
}
rf_rs = RandomizedSearchCV(
    RandomForestClassifier(random_state=42), rf_params, n_iter=20,
    scoring="f1_macro", cv=cv, random_state=42, n_jobs=-1
)
rf_rs.fit(X_train, y_train)
print("Best Random Forest Params:", rf_rs.best_params_)
print("Best Random Forest Score (CV F1):", rf_rs.best_score_)
```

# 7. Model Evaluation

â€¢ Evaluate models using:

--- Accuracy

--- Precision, Recall, F1-score

--- Confusion Matrix

--- Plot ROC Curve for the best-performing model.

```python
# Step 7.1. Evaluation of Tuned Models
# Weâ€™ll compare the tuned models (KNN, Decision Tree, Random Forest) on the **test set**, then plot the ROC curve for the best-performing one.
# Print metrics + confusion matrix for each tuned model and Create a comparison table.

Create a comparison table


from sklearn.metrics import precision_score, recall_score, f1_score, classification_report, ConfusionMatrixDisplay

def evaluate_model(name, model, X_test, y_test):
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average="macro", zero_division=0)
    rec = recall_score(y_test, y_pred, average="macro", zero_division=0)
    f1 = f1_score(y_test, y_pred, average="macro", zero_division=0)

    print(f"\n{name} Results:")
    print("Accuracy :", acc)
    print("Precision:", prec)
    print("Recall   :", rec)
    print("F1 Score :", f1)
    print("\nClassification Report:\n", classification_report(y_test, y_pred, zero_division=0))

    ConfusionMatrixDisplay.from_predictions(y_test, y_pred)

    return {"Model": name, "Accuracy": acc, "Precision": prec, "Recall": rec, "F1": f1}

# Evaluate all tuned models
results = []
results.append(evaluate_model("KNN (Tuned)", knn_rs.best_estimator_, X_test, y_test))
results.append(evaluate_model("Decision Tree (Tuned)", dt_rs.best_estimator_, X_test, y_test))
results.append(evaluate_model("Random Forest (Tuned)", rf_rs.best_estimator_, X_test, y_test))

# Put results in a DataFrame
results_df = pd.DataFrame(results).set_index("Model")
results_df
```

```python
# ROC Curve for the Best Model

from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt

# Pick the best model by F1 score
best_model_name = results_df["F1"].idxmax()
best_model = {
    "KNN (Tuned)": knn_rs.best_estimator_,
    "Decision Tree (Tuned)": dt_rs.best_estimator_,
    "Random Forest (Tuned)": rf_rs.best_estimator_
}[best_model_name]

print("Best model is:", best_model_name)

# ROC Curve (multiclass -> One-vs-Rest)
y_score = best_model.predict_proba(X_test)
y_bin = label_binarize(y_test, classes=["Low", "Medium", "High"])
n_classes = y_bin.shape[1]

plt.figure(figsize=(7,6))
for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_bin[:, i], y_score[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"Class {i} (AUC={roc_auc:.2f})")

plt.plot([0,1], [0,1], "k--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"ROC Curve â€” {best_model_name}")
plt.legend(loc="lower right")
plt.show()
```
